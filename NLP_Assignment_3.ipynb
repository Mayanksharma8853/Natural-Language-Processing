### Apply NER

```python
import spacy
nlp=spacy.load('en_core_web_sm')
```

```python
text="""Sundar Pichai is the Chief Executive Officer of Google and Alphabet. He was born in Chennai, India, and studied engineering at IIT Kharagpur. Sundar Pichai later pursued higher education in the United States and earned a master’s degree from Stanford University. Before joining Google, he worked in the semiconductor industry. At Google, Sundar Pichai played a key role in developing the Chrome browser. His leadership style emphasizes innovation, collaboration, and simplicity. Under his guidance, Google expanded its focus on artificial intelligence. Artificial intelligence is now integrated into many Google products. Sundar Pichai believes technology should benefit everyone and often speaks about the ethical use of artificial intelligence. Data privacy and user trust are important concerns for technology companies, and Sundar Pichai has addressed these issues in global technology conferences. His journey inspires many young professionals around the world, and leadership and responsibility remain central to his vision for the future."""
```

```python
doc=nlp(text)
```

```python
for ent in doc.ents:
    print(ent.text, "-", ent.label_)
```

```python
from spacy import displacy

displacy.render(doc,style='ent',jupyter=True)
```

```python
displacy.render(doc, style="dep", jupyter=True)
```
