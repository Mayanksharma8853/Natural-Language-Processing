```python
import spacy
from spacy import displacy

nlp = spacy.load("en_core_web_sm")

doc = nlp(text)

displacy.render(doc, style="dep", jupyter=True)
```

```python
doc = nlp(text)

for token in doc:
    print(
        f"Word: {token.text:15} "
        f"POS: {token.pos_:6} "
        f"Head: {token.head.text:15} "
        f"Dependency: {token.dep_}"
    )
```
